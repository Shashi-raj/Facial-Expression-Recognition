```python
from keras.models import model_from_json
import numpy as np
import matplotlib.pyplot as plt
from keras.utils import to_categorical
from keras.preprocessing.image import load_img
import pandas as pd
import os
import random
```

```python
json_file = open("Face_expression_model.json", "r")
model_json = json_file.read()
json_file.close()
model = model_from_json(model_json)
model.load_weights("Face_expression_model.h5")
```

```python
label = ['angry','disgust','fear','happy','neutral','sad','surprise']
```

```python
def ef(image):
    img = load_img(image, color_mode='grayscale', target_size=(150,150))
    feature = np.array(img)
    feature = feature.reshape(1,150,150,1)
    return feature/255.0
```

```python
image = 'images/Happy.jpg'
print("original image is of Happy")
img1 = load_img(image, color_mode='grayscale', target_size=(150,150))
img = ef(image)
pred = model.predict(img)
pred_label = label[pred.argmax()]
print("model prediction is ",pred_label)
img_arr = np.array(img1)
plt.imshow(img_arr, cmap='gray')
plt.axis('off')
plt.show()
```
